**1. IMPORT DATABASE AND PRINT OUT ALL THE ARTISTS AND THEIR ALBUMS**

```python
import sqlite3
from pathlib import Path
import pandas as pd

# Connect to the SQLite database
con = sqlite3.connect(Path("data/Chinook.sqlite"))

# Create a cursor object
cursor = con.cursor()

query = """
SELECT Artist.Name AS ArtistName, Album.Title AS AlbumTitle
FROM Artist
JOIN Album ON Artist.ArtistId = Album.ArtistId
ORDER BY ArtistName, AlbumTitle;"""

# Execute the query
cursor.execute(query)

# Fetch all rows
rows = cursor.fetchall()

# Print all the artists and their album title
for row in rows:
    artist_name, album_title = row
    print(f"Artist: {artist_name}, Album: {album_title}")


# Close the database connection
con.close()
```

**2. CREATE A NEW TABLE CALLED "TOPTENLONGESTTRACKS" IN THE DATABASE**

```python
# Connect to the SQLite database
con = sqlite3.connect(Path("data/Chinook.sqlite"))

# Create a cursor object
cursor = con.cursor()

# delete all the past TopTenLongestTracks
q1 = """
DROP TABLE TopTenLongestTracks;
"""

cursor.execute(q1)

create_new_table = """
CREATE TABLE IF NOT EXISTS TopTenLongestTracks (
    TrackId INTEGER,
    Name TEXT NO NULL,
    Milliseconds INTEGER
);
"""

# Execute the query1
cursor.execute(create_new_table)

# print out done
print("done")

# Close the database connection
con.close()
```

**3. PRINT OUT ALL THE TABLE NAMES IN THE DATABASE**

```python

# Connect to your SQLite database
conn = sqlite3.connect(Path("data/Chinook.sqlite"))  # Replace 'your_database.db' with your database file path

# Create a cursor object
cursor = conn.cursor()

# Query the sqlite_master table to retrieve table names
# Sqlite_master stores metadata about all tables in the database
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Fetch all table names
table_names = cursor.fetchall()

# Print the table names
for table in table_names:
    print(table[0])

# Close the cursor and the connection
cursor.close()
conn.close()
```

**5. ADD THE TOP 10 LONGEST TRACKS TO THE TABLE NAMED "TOPTENLONGESTTRACKS"**

```python
# Connect to the SQLite database
conn = sqlite3.connect(Path("data/Chinook.sqlite"))

# Create a cursor object
cursor = conn.cursor()

# Define an SQL query to fetch the 10 longest songs - descending
query = """
SELECT TrackId, Name, Milliseconds
FROM Track
ORDER BY Milliseconds DESC
LIMIT 10;
"""

# Execute the query
cursor.execute(query)

# Fetch the 10 longest songs
longest_songs = cursor.fetchall()

# Create the "toptenlongesttracks" table if it doesn't exist
create_new_table_query = """
CREATE TABLE IF NOT EXISTS TopTenLongestTracks (
    TrackId INTEGER,
    Name TEXT,
    Milliseconds INTEGER
);
"""

cursor.execute(create_new_table_query)

# Insert the fetched data into the "toptenlongesttracks" table
insert_data_query = """
INSERT INTO TopTenLongestTracks (TrackId, Name, Milliseconds)
VALUES (?, ?, ?);
"""

cursor.executemany(insert_data_query, longest_songs)

# Commit the transaction to save the changes
conn.commit()

# Close the database connection
conn.close()
```

**6. PRINT OUT THE CONTENT OF THE TABLE "TOPTENLONGESTTRACKS"**

```python
conn = sqlite3.connect(Path("data/Chinook.sqlite"))
cursor = conn.cursor()

# Select and print the content of the "toptenlongesttracks" table
select_query = """
SELECT *
FROM TopTenLongestTracks;
"""

cursor.execute(select_query)

# Fetch all rows from the table
table_data = cursor.fetchall()

# Convert the fetched data into a pandas DataFrame for easy printing
df = pd.DataFrame(table_data, columns=["TrackId", "Name", "Milliseconds"])

# Print the content of the "toptenlongesttracks" table
print("Contents of the 'toptenlongesttracks' table:")
print(df)
```
